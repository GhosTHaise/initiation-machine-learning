```python
dataset = {f'experience{i}' : np.random.randn(100,3) for i in range(4)}
dataset
```

```python
print("GhosT")
def graphique(dataset):
    n = len(dataset)
    plt.figure(figsize=(12,12))
    for i in range(n):
        plt.subplot(n,1,i+1)
        plt.plot(dataset[f'experience{i}'])
        plt.title(f'experience {i}')
    plt.show()    

graphique(dataset)

print("\n\nMachine Learnia")
def graphique1(dataset):
    n = len(dataset)
    plt.figure(figsize=(12,30))
    for k,i in zip (dataset.keys(),range(1,n+1)):
        plt.subplot(n,1,i)
        plt.plot(dataset[k])
        plt.title(k)
    plt.show()    

graphique1(dataset)
```
