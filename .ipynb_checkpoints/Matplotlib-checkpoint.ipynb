```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,2,10,dtype=np.float16)
y = x**2
print(x,y)
```

```python
plt.figure()
plt.plot(x,y,label='quadratique',c="green",lw=0.5,ls="--")
plt.plot(x,x**3,label="cubique")
plt.title("Linear graph")
plt.xlabel("axis x")
plt.ylabel("axis y")
plt.legend()


plt.savefig("test2.png");
plt.show()


print("\t------------------------")
```
