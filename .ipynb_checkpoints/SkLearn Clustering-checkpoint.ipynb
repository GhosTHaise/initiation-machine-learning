```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt 
import numpy as np
```

```python
X , y = make_blobs(n_samples=100,centers=3)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
model = KMeans(n_clusters=3)
model.fit(X)
```

```python
model.inertia_
```

```python
plt.scatter(X[:,0],X[:,1],c=model.predict(X))
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],c="r")
```

```python
model.cluster_centers_
```

```python
inertia = []
k_range = range(1,20)

for k in k_range:
    type(k)
    model = KMeans(n_clusters=k).fit(X)
    inertia.append(model.inertia_)
    
```

```python
plt.plot(k_range,inertia)
plt.xticks(k_range, k_range)

plt.show()
```

```python
model_test = KMeans(n_clusters=2)
model_test.fit(X)


test = np.array([[-0.69369056,  7.37453364]])
model_test.predict(test).shape
```

```python
plt.scatter(X[:,0],X[:,1],c=model_test.predict(X))
plt.scatter(model_test.cluster_centers_[:,0],model_test.cluster_centers_[:,1],c="r")
plt.scatter(test[:,0],test[:,1],c=model_test.predict(test))
```

```python
from sklearn.decomposition import PCA
```

```python
X.shape
```

```python
modelPCA = PCA(n_components=2)
X_reduced = modelPCA.fit_transform(X)
```

```python
print(X_reduced.shape)
plt.scatter(X_reduced[:,0],X_reduced[:,1],c=y)
plt.colorbar()
```
